```python

def plot_line_chart_horizontal(df, title, x_label, y_label):
    df_sorted = df.sort_values(by='Resource Demand')
    
    plt.figure(figsize=(10, 6))

    plt.plot(df_sorted['Resource Demand'], df_sorted['Resource Capacity'], marker='s', color='blue', label='Resource Capacity')

    can_handle_df = df_sorted[df_sorted['Can Handle']]
    plt.scatter(can_handle_df['Resource Demand'], can_handle_df['Resource Capacity'], color='green', edgecolor='k', label='Can Handle', zorder=5)

    cannot_handle_df = df_sorted[~df_sorted['Can Handle']]
    plt.scatter(cannot_handle_df['Resource Demand'], cannot_handle_df['Resource Capacity'], color='red', edgecolor='k', label='Cannot Handle', zorder=5)

    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.xticks(range(0, int(df_sorted['Resource Demand'].max()) + 1))
    plt.yticks(range(0, int(df_sorted['Resource Capacity'].max()) + 1)) 
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def plot_line_chart_vertical(df, title, x_label, y_label):
    # Sort the DataFrame by 'Resource Demand'
    df_sorted = df.sort_values(by='Resource Demand')
    
    # Plotting setup
    plt.figure(figsize=(10, 6))

    # Plot the line for 'Resource Capacity'
    plt.plot(df_sorted['Resource Demand'], df_sorted['Resource Capacity'], marker='s', linestyle='-', color='blue', label='Resource Capacity')

    # Scatter plot for 'Can Handle' = True
    can_handle_df = df_sorted[df_sorted['Can Handle']]
    plt.scatter(can_handle_df['Resource Demand'], can_handle_df['Resource Capacity'], color='green', edgecolor='k', label='Can Handle', zorder=5)

    # Scatter plot for 'Can Handle' = False
    cannot_handle_df = df_sorted[~df_sorted['Can Handle']]
    plt.scatter(cannot_handle_df['Resource Demand'], cannot_handle_df['Resource Capacity'], color='red', edgecolor='k', label='Cannot Handle', zorder=5)


    # Graph labels and legend
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.xticks(range(int(df_sorted['Resource Demand'].min()), int(df_sorted['Resource Demand'].max()) + 1))
    plt.yticks(range(0, int(df_sorted['Resource Capacity'].max()) + 1, max(1, int(df_sorted['Resource Capacity'].max()/10))))
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
plot_title = 'Horizontal Scaling Assessment with UI Latency SLO (95th Percentile <= 200ms) with Load Balancer'
plot_line_chart_horizontal(df1, plot_title, 'Number of Concurrent Users', 'Number of Pods')
```
